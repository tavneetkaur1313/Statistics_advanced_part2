2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans - The null hypothesis and alternative hypothesis are two statements used in statistical testing to explore a research question. The null hypothesis (H0) proposes that there is no significant effect or relationship between variables being studied, while the alternative hypothesis (H1 or Ha) suggests that a significant effect or relationship does exist. The goal of hypothesis testing is to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

Null Hypothesis  (H0)

-> Definition - Statement of no effect, relationship, or difference.

-> Purpose - Assumes no significant difference.

-> Testing Goal - Reject or fail to reject.

Alternative Hypothesis (H1/Ha)

-> Definition - Statement that contradicts H0, suggesting an effect, relationship, or difference.

-> Purpose - Asserts a significant difference exists.

-> Testing Goal - Support or accept.

3.  What is the significance level in hypothesis testing, and why is it important?

Ans - In hypothesis testing, the significance level (alpha, denoted as α) is the probability of incorrectly rejecting the null hypothesis when it's actually true. It's essentially the threshold for determining statistical significance and is a crucial factor in deciding whether to accept or reject the null hypothesis. Significance levels, like 0.05 (5%), are pre-determined before data collection, along with the potential outcomes of errors.


Why it's important?

-> Controlling for Type I errors: By setting α, researchers can control the likelihood of falsely concluding that there's a real effect when there isn't one.

-> Establishing a threshold for significance: It provides a clear criterion for determining whether the results are statistically significant, meaning they're unlikely to have occurred by chance.

-> Enabling comparison of results: Different studies can be compared based on their significance levels and p-values, allowing for a more robust evaluation of findings.

4.  What does a P-value represent in hypothesis testing?

It essentially quantifies the strength of evidence against the null hypothesis. A smaller p-value indicates stronger evidence against the null hypothesis, suggesting that the observed results are unlikely to have occurred by chance if the null hypothesis were true.


 A p-value is a statistical measurement used to validate a hypothesis against observed data. A p-value measures the probability of obtaining the observed results, assuming that the null hypothesis is true. The lower the p-value, the greater the statistical significance of the observed difference.

5.  How do you interpret the P-value in hypothesis testing?

Ans - In hypothesis testing, the p-value represents the probability of observing a result as extreme as (or more extreme than) the one obtained in the study, assuming the null hypothesis is true. Essentially, it quantifies how unlikely the observed results are if the null hypothesis holds true. A smaller p-value provides stronger evidence against the null hypothesis, suggesting the observed effect is statistically significant.

How to interpret the p-value?


-> Compare to the significance level (alpha):

A commonly used significance level (alpha) is 0.05 (5%). This means that if the p-value is less than or equal to 0.05, the observed result is considered statistically significant.


-> Reject or fail to reject the null hypothesis:

If p ≤ alpha (e.g., p < 0.05): You reject the null hypothesis, meaning there is strong evidence that the observed effect is not due to chance alone.

If p > alpha (e.g., p > 0.05): You fail to reject the null hypothesis, meaning there is not enough evidence to conclude that the observed effect is statistically significant. This doesn't necessarily mean the null hypothesis is true, but rather that the data doesn't provide strong enough evidence to reject it.

-> Strength of evidence:

A smaller p-value (e.g., p < 0.01) indicates stronger evidence against the null hypothesis, while a larger p-value (e.g., p > 0.05) indicates weaker evidence.


6. What are Type 1 and Type 2 errors in hypothesis testing?

Ans -
In hypothesis testing, a Type I error (false positive) occurs when you reject a true null hypothesis, and a Type II error (false negative) occurs when you fail to reject a false null hypothesis.


1. Type I Error (False Positive):

-> Definition: This error happens when you reject the null hypothesis (H₀) when it is actually true. In other words, you conclude there's an effect or relationship when there isn't one.

-> Example: Imagine you are testing a new drug for a disease. You find statistically significant evidence that the drug is effective, but in reality, the drug has no effect. This would be a Type I error.

-> Symbol: The probability of a Type I error is often denoted by α (alpha).
Also known as: False alarm, false positive.

2. Type II Error (False Negative):

-> Definition: This error occurs when you fail to reject the null hypothesis when it is actually false. You conclude that there's no effect or relationship when one actually exists.

-> Example: Using the same drug example, you might not find statistically significant evidence of its effectiveness, even though it actually is. This would be a Type II error.

-> Symbol: The probability of a Type II error is often denoted by β (beta).
Also known as: Missed detection, false negative.

-> Power: The probability of correctly rejecting a false null hypothesis is called the power of the test. It's related to Type II error as power = 1 - β.

9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?

Ans - The z-score, also known as a standard score, quantifies how many standard deviations a data point is from the mean of a distribution. In hypothesis testing, it helps determine the probability of observing a sample mean (or other statistic) if the null hypothesis is true. The Z-score is calculated using the formula: Z = (x - μ) / σ, where 'x' is the data point, 'μ' is the population mean, and 'σ' is the population standard deviation.

1. Calculation:

-> Identify the data point (x): This is the specific value you want to standardize.

-> Determine the population mean (μ) and standard deviation (σ): You need to know the average and spread of the population you are interested in.

-> Subtract the mean from the data point (x - μ): This gives the difference between the data point and the average.

-> Divide the result by the standard deviation ( (x - μ) / σ ): This final step gives the z-score.

2. Representation in Hypothesis Testing:

-> Null Hypothesis:
The z-score helps determine whether the observed sample mean is significantly different from the hypothesized population mean, assuming the null hypothesis is true.

-> P-value:
The z-score is used to calculate the p-value, which represents the probability of obtaining a sample mean as extreme as (or more extreme than) the observed sample mean, assuming the null hypothesis is true.

-> Decision:
If the p-value is below a chosen significance level (alpha), the null hypothesis is rejected, suggesting that the observed sample mean is not likely due to chance alone.

12. What is the T-test, and how is it used in hypothesis testing?

Ans - A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups. It's used in hypothesis testing to compare the means of two samples and decide if the observed difference is due to chance or a real effect.

1. Defining the Hypotheses:

Null Hypothesis (H0): There is no significant difference between the means of the two groups. In other words, the observed difference is due to random chance.

Alternative Hypothesis (H1): There is a significant difference between the means of the two groups.

2. Conducting the T-test:

The t-test calculates a test statistic (t-value) based on the sample data, taking into account the difference in means and the variability within each group.

This t-value is then compared to a critical value based on the chosen significance level (usually 0.05) and the degrees of freedom.

3. Interpreting the Results:

If the calculated t-value is greater than the critical value:

You reject the null hypothesis, suggesting that there is a statistically significant difference between the means.

If the calculated t-value is less than or equal to the critical value:
You fail to reject the null hypothesis, meaning that the difference between the means may be due to chance.

4. Choosing the Right T-test:

-> Independent Samples T-test:

Used when you have two separate groups, and the observations in one group are not related to the observations in the other group.

-> Paired Samples T-test:

Used when you have two related groups (e.g., before-and-after measurements on the same individuals).

-> One-Sample T-test:

Used to compare the mean of a single group to a known or hypothesized value.

16. How is Bayes' Theorem used in statistics, and what is its significance?

Ans - Bayes' Theorem is a fundamental concept in probability and statistics used to update the probability of a hypothesis when new evidence is available. It allows us to revise our beliefs about an event based on new information, effectively bridging prior knowledge with new observations. The theorem's significance lies in its ability to provide a framework for rational decision-making under uncertainty, especially in fields like machine learning, healthcare, and finance.

Here's how it's used and its significance:

-> Conditional Probability:

Bayes' Theorem calculates the conditional probability of an event (A) given the occurrence of another event (B), or P(A|B). It essentially answers: How does the probability of A change when we know that B has happened?

-> Updating Probabilities:

The theorem allows us to take a prior probability (our initial belief about A) and refine it using new evidence (B) to arrive at a posterior probability. This posterior probability represents our updated belief about A, incorporating the new evidence.

-> Inference and Decision-Making:

By updating probabilities, Bayes' Theorem helps us make more informed decisions, especially in situations where the outcome is uncertain. For example, in medical diagnosis, it helps determine the probability of a disease given the presence of certain symptoms.

-> Machine Learning:
Bayes' Theorem is a cornerstone of many machine learning algorithms, particularly in classification tasks like spam detection, text analysis, and medical diagnosis. It's the basis for the Naive Bayes classifier.

-> Bayesian Statistics:

Bayesian statistics is an entire branch of statistical inference that relies heavily on Bayes' Theorem. It uses subjective probabilities (prior beliefs) and incorporates them with new data to draw conclusions.

18.  What is the Chi-square goodness of fit test, and how is it applied?


Ans - The Chi-square goodness of fit test determines how well a sample's observed distribution aligns with a hypothesized population distribution. It's a non-parametric test used to compare observed frequencies in categories to expected frequencies based on a theoretical model. The test statistic, calculated using observed and expected frequencies, is compared to a Chi-square distribution to determine if the observed data significantly deviates from the expected.


How it's applied

1. State the Hypothesis:

Define the null hypothesis (the population follows a specific distribution) and the alternative hypothesis (the population does not follow that distribution).

2. Calculate Expected Frequencies:

Determine the expected number of observations in each category based on the hypothesized distribution.

3. Calculate the Test Statistic:

Use the following formula: Q² = ∑(Observed Frequency - Expected Frequency)² / Expected Frequency.

4. Determine Degrees of Freedom:

5. Find the Critical Value:

Using the degrees of freedom and the chosen significance level (α), find the critical value from the Chi-square distribution table or use a statistical software package.

6. Compare and Decide:

Compare the calculated test statistic with the critical value. If the test statistic is greater than the critical value, reject the null hypothesis, suggesting the sample does not fit the hypothesized distribution.

20.  What is an ANOVA test, and what are its assumptions?

Ans - ANOVA (Analysis of Variance) is a statistical test used to compare the means of two or more groups. Its assumptions are: normality, homogeneity of variance, and independent observations.

1. Normality: The data within each group should be approximately normally distributed. This means the distribution should resemble a bell curve, with most values clustered around the mean. While a perfectly normal distribution is not always necessary, significant deviations (like extreme skewness or outliers) can affect the reliability of the ANOVA results, especially with small sample sizes.

2. Homogeneity of Variance: The variance (spread or variability) within each group should be roughly equal. This means the data should not have vastly different spread between groups. If the variances are significantly different, it violates the assumption of homogeneity of variance, also known as homoscedasticity.

3. Independent Observations: Each observation within a group should be independent of the others. This means that the value of one data point should not influence the value of another. For example, if you're measuring the effectiveness of a drug on a group of patients, each patient's response should not depend on another patient's response.

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(p, x):
    """
    Calculate the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (0 <= p <= 1)
    x (int): The outcome (0 or 1)

    Returns:
    float: The probability of the outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1.")
    if p < 0 or p > 1:
        raise ValueError("p must be between 0 and 1.")

    return p if x == 1 else (1 - p)

# Example usage
p = 0.7  # Probability of success
outcomes = [0, 1]

for x in outcomes:
    probability = bernoulli_pdf(p, x)
    print(f"P(X = {x}) = {probability}")
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
a = 1  # minimum value
b = 6  # maximum value

# Generate the possible outcomes
outcomes = np.arange(a, b + 1)

# Calculate the probabilities for each outcome
probabilities = np.full(outcomes.shape, 1 / len(outcomes))

# Calculate the CDF
cdf = np.cumsum(probabilities)

# Plotting the CDF
plt.figure(figsize=(10, 6))
plt.step(outcomes, cdf, where='post', color='blue', label='CDF', linewidth=2)

# Adding titles and labels
plt.title('Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution')
plt.xlabel('Outcomes')
plt.ylabel('Cumulative Probability')
plt.xticks(outcomes)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(axis='y')
plt.legend()

# Show the plot
plt.show()
```

16.  Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a dataset.

    Parameters:
        data (list or np.array): Sample data points.
        confidence (float): Confidence level (default 0.95).

    Returns:
        tuple: Lower bound and upper bound of confidence interval.
    """
    n = len(data)                      # Sample size
    mean = np.mean(data)               # Sample mean
    std_err = stats.sem(data)          # Standard error of the mean
    t_score = stats.t.ppf((1 + confidence) / 2, n - 1)  # t critical value

    margin_of_error = t_score * std_err

    return mean - margin_of_error, mean + margin_of_error

# Sample dataset
data = [12, 15, 14, 10, 13, 16, 14, 15, 11, 13]

# Calculate confidence interval
conf_level = 0.95
lower_bound, upper_bound = confidence_interval(data, conf_level)

# Print results
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"{conf_level*100:.1f}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation
print(f"\nInterpretation: We are {conf_level*100:.1f}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
import scipy.stats as stats

def z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05):
    """
    Perform a Z-test to compare a sample mean to a known population mean.

    Parameters:
    - sample_mean: Mean of the sample.
    - population_mean: Known population mean.
    - population_std_dev: Known population standard deviation.
    - sample_size: Size of the sample.
    - alpha: Significance level (default is 0.05).

    Returns:
    - z_score: Calculated Z-score.
    - p_value: P-value for the test.
    - result: Interpretation of the test result.
    """
    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Calculate the p-value
    p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))  # Two-tailed test

    # Determine the critical value
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Interpret the result
    if np.abs(z_score) > z_critical:
        result = "Reject the null hypothesis: There is a significant difference."
    else:
        result = "Fail to reject the null hypothesis: No significant difference."

    return z_score, p_value, result

# Example usage
sample_mean = 11.8
population_mean = 12
population_std_dev = 0.5
sample_size = 100

z_score, p_value, result = z_test(sample_mean, population_mean, population_std_dev, sample_size)

print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")
print(result)
```

4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05):
    """
    Perform a two-tailed Z-test and visualize decision regions.

    Parameters:
    - sample_mean: Mean of the sample.
    - population_mean: Known population mean.
    - population_std_dev: Known population standard deviation.
    - sample_size: Size of the sample.
    - alpha: Significance level (default 0.05).

    Returns:
    - z_score: Calculated Z-score.
    - p_value: P-value for the test.
    - decision: "Reject" or "Fail to reject" null hypothesis.
    """
    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Critical values for rejection regions
    z_critical = stats.norm.ppf(1 - alpha/2)

    # Decision
    if abs(z_score) > z_critical:
        decision = "Reject the null hypothesis: significant difference."
    else:
        decision = "Fail to reject the null hypothesis: no significant difference."

    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.5)

    # Shade acceptance region
    plt.fill_between(x, 0, y, where=((x > -z_critical) & (x < z_critical)), color='green', alpha=0.3, label='Acceptance Region')

    # Plot the calculated Z-score
    plt.axvline(z_score, color='black', linestyle='--', label=f'Z-Score = {z_score:.2f}')

    # Titles and labels
    plt.title('Two-Tailed Z-Test and Decision Regions')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_score, p_value, decision

# Example usage:
sample_mean = 105
population_mean = 100
population_std_dev = 15
sample_size = 30
alpha = 0.05

z_score, p_value, decision = two_tailed_z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha)

print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")
print(decision)
```

5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(
    mu_null,
    mu_alt,
    sigma,
    n,
    alpha=0.05,
    alternative='two-sided'  # can be 'two-sided', 'less', or 'greater'
):
    """
    Calculate and visualize Type 1 and Type 2 errors for hypothesis testing.

    Parameters:
    - mu_null: Population mean under null hypothesis (H0)
    - mu_alt: True population mean under alternative hypothesis (H1)
    - sigma: Population standard deviation (assumed known)
    - n: Sample size
    - alpha: Significance level
    - alternative: Type of test ('two-sided', 'less', 'greater')

    Returns:
    - type1_error: Probability of Type 1 error (α)
    - type2_error: Probability of Type 2 error (β)
    """

    # Standard error of the mean
    se = sigma / np.sqrt(n)

    # Calculate critical values based on alpha and alternative hypothesis
    if alternative == 'two-sided':
        z_crit_low = stats.norm.ppf(alpha / 2)
        z_crit_high = stats.norm.ppf(1 - alpha / 2)
        crit_low = mu_null + z_crit_low * se
        crit_high = mu_null + z_crit_high * se

        # Type 1 error region: rejection regions beyond critical values under H0
        # (already defined by alpha)

        # Type 2 error: probability of accepting H0 when H1 is true
        # It is the probability that the sample mean lies between crit_low and crit_high under H1
        beta = stats.norm.cdf(crit_high, loc=mu_alt, scale=se) - stats.norm.cdf(crit_low, loc=mu_alt, scale=se)

        rejection_regions = [(None, crit_low), (crit_high, None)]
        acceptance_region = (crit_low, crit_high)

    elif alternative == 'less':
        z_crit = stats.norm.ppf(alpha)
        crit_val = mu_null + z_crit * se

        # Type 2 error: probability sample mean > crit_val under H1 (fail to reject H0 when H1 true)
        beta = 1 - stats.norm.cdf(crit_val, loc=mu_alt, scale=se)

        rejection_regions = [(None, crit_val)]
        acceptance_region = (crit_val, None)

    elif alternative == 'greater':
        z_crit = stats.norm.ppf(1 - alpha)
        crit_val = mu_null + z_crit * se

        # Type 2 error: probability sample mean < crit_val under H1
        beta = stats.norm.cdf(crit_val, loc=mu_alt, scale=se)

        rejection_regions = [(crit_val, None)]
        acceptance_region = (None, crit_val)

    else:
        raise ValueError("Invalid alternative hypothesis. Choose from 'two-sided', 'less', or 'greater'.")

    # Type 1 error is alpha
    alpha_val = alpha

    # Plotting
    x_min = min(mu_null, mu_alt) - 4 * se
    x_max = max(mu_null, mu_alt) + 4 * se
    x = np.linspace(x_min, x_max, 1000)

    # Distributions under H0 and H1
    y_null = stats.norm.pdf(x, mu_null, se)
    y_alt = stats.norm.pdf(x, mu_alt, se)

    plt.figure(figsize=(12,6))

    # Plot distributions
    plt.plot(x, y_null, label='Null Hypothesis (H0) Distribution', color='blue')
    plt.plot(x, y_alt, label='Alternative Hypothesis (H1) Distribution', color='orange')

    # Shade rejection regions (Type 1 error)
    for region in rejection_regions:
        low, high = region
        if low is None:
            xs = x[x <= high]
            ys = y_null[x <= high]
        elif high is None:
            xs = x[x >= low]
            ys = y_null[x >= low]
        else:
            xs = x[(x >= low) & (x <= high)]
            ys = y_null[(x >= low) & (x <= high)]
        plt.fill_between(xs, 0, ys, color='red', alpha=0.4, label='Type 1 Error Region' if region == rejection_regions[0] else None)

    # Shade acceptance region under H1 (Type 2 error)
    low, high = acceptance_region
    if low is None:
        xs = x[x <= high]
        ys = y_alt[x <= high]
    elif high is None:
        xs = x[x >= low]
        ys = y_alt[x >= low]
    else:
        xs = x[(x >= low) & (x <= high)]
        ys = y_alt[(x >= low) & (x <= high)]
    plt.fill_between(xs, 0, ys, color='green', alpha=0.4, label='Type 2 Error Region')

    # Labels and legend
    plt.title('Type 1 (α) and Type 2 (β) Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Type 1 Error (α): {alpha_val:.4f}")
    print(f"Type 2 Error (β): {beta:.4f}")

    return alpha_val, beta

# Example usage:
# Null hypothesis mean = 100
# True mean = 105
# Population std dev = 15
# Sample size = 30
# Significance level = 0.05

visualize_type1_type2_errors(mu_null=100, mu_alt=105, sigma=15, n=30, alpha=0.05, alternative='two-sided')
```

6. Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def independent_t_test(group1, group2, alpha=0.05):
    """
    Perform an independent T-test and interpret the results.

    Parameters:
    - group1: Array-like, the first group of data.
    - group2: Array-like, the second group of data.
    - alpha: Significance level (default is 0.05).

    Returns:
    - t_stat: T-statistic.
    - p_value: P-value for the test.
    - result: Interpretation of the test result.
    """
    # Perform the independent T-test
    t_stat, p_value = stats.ttest_ind(group1, group2)

    # Interpret the result
    if p_value < alpha:
        result = "Reject the null hypothesis: There is a significant difference between the groups."
    else:
        result = "Fail to reject the null hypothesis: No significant difference between the groups."

    return t_stat, p_value, result

# Example usage
# Simulated data for two groups
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=100, scale=10, size=30)  # Group 1: mean=100, std=10
group2 = np.random.normal(loc=110, scale=15, size=30)  # Group 2: mean=110, std=15

# Perform the independent T-test
t_stat, p_value, result = independent_t_test(group1, group2)

# Print the results
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print(result)

# Optional: Visualize the data
plt.figure(figsize=(10, 6))
plt.boxplot([group1, group2], labels=['Group 1', 'Group 2'])
plt.title('Boxplot of Two Independent Groups')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

7.  Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(mid_grades, final_grades):
    """
    Perform a paired T-test and visualize the results.

    Parameters:
    - mid_grades: Array-like, the midterm grades.
    - final_grades: Array-like, the final grades.

    Returns:
    - t_stat: T-statistic.
    - p_value: P-value for the test.
    """
    # Calculate the differences
    differences = final_grades - mid_grades

    # Perform the paired T-test
    t_stat, p_value = stats.ttest_rel(mid_grades, final_grades)

    # Visualize the differences
    plt.figure(figsize=(10, 6))
    plt.hist(differences, bins=10, alpha=0.7, color='blue', edgecolor='black')
    plt.axvline(x=np.mean(differences), color='red', linestyle='dashed', linewidth=1)
    plt.text(np.mean(differences) + 1, 5, f'Mean: {np.mean(differences):.2f}', color='red')
    plt.title('Histogram of Differences between Midterm and Final Grades')
    plt.xlabel('Differences (Final - Midterm)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    return t_stat, p_value

# Example usage
# Simulated data for midterm and final grades
np.random.seed(42)  # For reproducibility
mid_grades = np.random.normal(loc=75, scale=10, size=30)  # Midterm grades
final_grades = mid_grades + np.random.normal(loc=5, scale=5, size=30)  # Final grades with a slight increase

# Perform the paired T-test
t_stat, p_value = paired_t_test(mid_grades, final_grades)

# Print the results
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation of results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between midterm and final grades.")
else:
    print("Fail to reject the null hypothesis: No significant difference between midterm and final grades.")
```

8.  Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
import scipy.stats as stats

def one_sample_z_test(sample, population_mean, population_std):
    """
    Perform a one-sample Z-test comparing sample mean to population mean.
    """
    sample_mean = np.mean(sample)
    n = len(sample)
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed
    return z_stat, p_value

def one_sample_t_test(sample, population_mean):
    """
    Perform a one-sample T-test comparing sample mean to population mean.
    """
    t_stat, p_value = stats.ttest_1samp(sample, population_mean)
    return t_stat, p_value

# Simulate sample data
np.random.seed(42)
population_mean = 50
population_std = 10  # Known population standard deviation
sample_size = 30

# Generate sample data from normal distribution
sample_data = np.random.normal(loc=52, scale=population_std, size=sample_size)  # Slightly different mean

# Perform Z-test (using known population std)
z_stat, z_p_value = one_sample_z_test(sample_data, population_mean, population_std)

# Perform T-test (using sample std)
t_stat, t_p_value = one_sample_t_test(sample_data, population_mean)

# Print results
print("One-sample Z-test results:")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value = {z_p_value:.4f}\n")

print("One-sample T-test results:")
print(f"T-statistic = {t_stat:.4f}")
print(f"P-value = {t_p_value:.4f}\n")

# Interpretation
alpha = 0.05
print("Interpretation at alpha = 0.05:")
print(f"Z-test: {'Reject' if z_p_value < alpha else 'Fail to reject'} the null hypothesis.")
print(f"T-test: {'Reject' if t_p_value < alpha else 'Fail to reject'} the null hypothesis.")
```

9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    data (list or np.array): Sample data
    confidence (float): Confidence level (default is 0.95 for 95% CI)

    Returns:
    tuple: Lower and upper bounds of the confidence interval
    """
    # Convert data to a numpy array for calculations
    data = np.array(data)

    # Calculate the sample mean and standard error
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)  # Sample size
    standard_error = sample_std / np.sqrt(n)

    # Calculate the critical t-value
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Calculate the margin of error
    margin_of_error = t_critical * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = [2.3, 2.5, 2.1, 2.4, 2.6]
ci = confidence_interval(sample_data)
print(f"Confidence Interval: {ci}")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
    data (list or np.array): Sample data
    confidence (float): Confidence level (default is 0.95 for 95% CI)

    Returns:
    float: Margin of error
    """
    # Convert data to a numpy array for calculations
    data = np.array(data)

    # Calculate the sample standard deviation and sample size
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)  # Sample size

    # Calculate the standard error
    standard_error = sample_std / np.sqrt(n)

    # Calculate the critical t-value
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Calculate the margin of error
    margin_of_error = t_critical * standard_error

    return margin_of_error

# Example usage
sample_data = [2.3, 2.5, 2.1, 2.4, 2.6]
confidence_level = 0.95
moe = margin_of_error(sample_data, confidence_level)
print(f"Margin of Error at {confidence_level*100}% confidence level: {moe:.4f}")
```

11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayesian_inference(prior_disease, likelihood_positive_given_disease, false_positive_rate):
    # Calculate the probability of not having the disease
    prior_no_disease = 1 - prior_disease

    # Calculate the marginal likelihood P(T)
    p_positive_test = (likelihood_positive_given_disease * prior_disease) + (false_positive_rate * prior_no_disease)

    # Calculate the posterior probability P(D|T)
    posterior_probability = (likelihood_positive_given_disease * prior_disease) / p_positive_test

    return posterior_probability

# Define the probabilities
P_D = 0.01  # Prior probability of disease
P_T_given_D = 0.99  # Likelihood of positive test given disease
P_T_given_not_D = 0.05  # Probability of positive test given no disease

# Calculate the posterior probability
posterior = bayesian_inference(P_D, P_T_given_D, P_T_given_not_D)
print(f"The probability of having the disease given a positive test result is {posterior:.2f}")
```

13.  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate the expected frequencies for a Chi-square test.

    Parameters:
    observed (np.array or pd.DataFrame): Observed frequency table

    Returns:
    np.array: Expected frequency table
    """
    # Convert to DataFrame if it's not already
    if isinstance(observed, np.ndarray):
        observed = pd.DataFrame(observed)

    # Calculate row totals, column totals, and grand total
    row_totals = observed.sum(axis=1).values.reshape(-1, 1)  # Reshape for broadcasting
    column_totals = observed.sum(axis=0).values.reshape(1, -1)  # Reshape for broadcasting
    grand_total = observed.values.sum()

    # Calculate expected frequencies
    expected = (row_totals @ column_totals) / grand_total
    return expected

# Example observed data (contingency table)
observed_data = np.array([[30, 10],  # Male preferences
                          [20, 40]]) # Female preferences

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Display results
print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
import scipy.stats as stats

def goodness_of_fit_test(observed, expected):
    """
    Perform a Chi-square goodness-of-fit test.

    Parameters:
    observed (list or np.array): Observed frequencies
    expected (list or np.array): Expected frequencies

    Returns:
    chi2_stat (float): Chi-square statistic
    p_value (float): p-value of the test
    """
    # Calculate the Chi-square statistic
    chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
    return chi2_stat, p_value

# Example observed data (e.g., counts of outcomes)
observed_data = np.array([50, 30, 20])  # Observed frequencies for three categories

# Example expected data (e.g., expected frequencies based on a theoretical distribution)
expected_data = np.array([40, 40, 20])  # Expected frequencies for the same categories

# Perform the goodness-of-fit test
chi2_stat, p_value = goodness_of_fit_test(observed_data, expected_data)

# Display results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"p-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```
